### In which country do you currently reside?

```python
df_location = df_response['Q4']
df_location.value_counts()
```

```python
df_location.isna().sum() # checking for missing values
```

##### NOTE: countries with very few respondents were grouped under "other" and because I want to classify countries into continents, I will have to take out "other" and "I do not wish to disclose my location"

```python
df_not_other = df_response[df_response['Q4'] != 'Other']
df_not_disclose = df_not_other[df_not_other['Q4'] != 'I do not wish to disclose my location']

df_country = df_not_disclose
```

```python
# I also have to rename some countries so pycountry library can understand it

df_country['Q4'] = df_country['Q4'].replace({
                                        'Iran, Islamic Republic of...': 'Iran',
                                        'Hong Kong (S.A.R.)': 'Hong Kong'})
```

```python
df_country.Q4.value_counts()
```

##### NOTE: numbers of countries are too much to visualize in a chart so I need to split them into continents for better visualization experience (there are just 6 continents so the visualization will be easy to comprehend)
#### Function for converting country to continent code

```python
def convert(row):
    country_code = pc.country_name_to_country_alpha2(row.Q4, cn_name_format='default')
    continent_code = pc.country_alpha2_to_continent_code(country_code)
    return continent_code
```

#### Create a column for continent code 

```python
df_country['continent'] = df_country.apply(convert, axis=1)
df_country.head()
```

#### Map continent code to continent names

```python
continent_names = {
                    'AS': 'Asia',
                    'AF': 'Africa',
                    'SA': 'South America',
                    'NA': 'North America',
                    'EU': 'Europe',
                    'OC': 'Australia'
}

df_country['continent'] = df_country['continent'].map(continent_names)
df_country.head()
```

```python
charts.bar_plot(df_country['continent'].value_counts().sort_values(), 'Exploring respondents based on continents')
```
